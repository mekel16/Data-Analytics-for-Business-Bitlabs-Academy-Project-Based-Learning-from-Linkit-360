```python
# columns = ['revenue', 'service_price', 'attempt_charging', 'success_billing']

for col in columns:
    percentile25 = data[col].quantile(0.25)
    percentile75 = data[col].quantile(0.75)
    iqr = percentile75 - percentile25

    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr

    print(f"Lower limit for {col}:", lower_limit)
    print(f"Upper limit for {col}:", upper_limit)

    outliers = data[(data[col] > upper_limit) | (data[col] < lower_limit)]

    print(f"Number of rows in the data containing outliers in {col}:", len(outliers))
    print()
```

EDA untuk churn risk score, pertama kita scale dulu kolom yang pelu di scaling yaitu attempt_charging dan success_billing


```python
def normalize(series, min_scale=0, max_scale=10):
    min_val = series.min()
    max_val = series.max()
    return min_scale + (series - min_val) * (max_scale - min_scale) / (max_val - min_val)

# Normalisasi kolom 'charging_attempts' dan 'successful_billing'
data['CA_normalized'] = normalize(data['attempt_charging'], min_scale=0, max_scale=10)
data['SB_normalized'] = normalize(data['success_billing'], min_scale=0, max_scale=10)

# Tampilkan hasil
data[['id', 'attempt_charging', 'CA_normalized', 'success_billing', 'SB_normalized']]
```

```python
import pandas as pd
# buat churn jika -1 selain itu bukan churn
churned_customers = data[data['subscription_status'] == -1].shape[0]
# Menghitung jumlah pelanggan pada awal periode
total_customers = data.shape[0]

# Menghitung churn rate
churn_rate = (churned_customers / total_customers) * 100

print(f"Churn Rate: {churn_rate:.2f}%")
```

```python
# Pemetaan subscription status ke skor
status_scores = {1: 10, 0: 5, -1: 0}
data['SS_Score'] = data['status'].map(status_scores)

# Hitung Engagement Score dengan subscription status
weights = {'CA': 0.4, 'SB': 0.4, 'SS': 0.2}
data['Engagement_Score'] = (
    weights['CA'] * data['CA_normalized'] +
    weights['SB'] * data['SB_normalized'] +
    weights['SS'] * data['SS_Score']
)

# Hitung churn risk
ES_min = data['Engagement_Score'].min()
ES_max = data['Engagement_Score'].max()
data['Churn_Risk'] = 1 - (data['Engagement_Score'] - ES_min) / (ES_max - ES_min)

# Tampilkan hasil
data[['id', 'status', 'Engagement_Score', 'Churn_Risk']]
```
