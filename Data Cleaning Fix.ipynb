```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler

file_path = '/content/drive/MyDrive/MSIB Batch 7/Project Based Learning/dataset_pbl.csv'
df = pd.read_csv(file_path)
data = pd.read_csv(file_path)
```

```python
print(data.isnull().sum())

# Fill missing values for 'adnet' with 'NA'
data['adnet'].fillna('NA', inplace=True)

# Convert date columns to datetime
data['subs_date'] = pd.to_datetime(data['subs_date'])
data['renewal_date'] = pd.to_datetime(data['renewal_date'], errors='coerce')
data['freemium_end_date'] = pd.to_datetime(data['freemium_end_date'], errors='coerce')
data['unsubs_date'] = pd.to_datetime(data['unsubs_date'], errors='coerce')

# Create derived features
data['sales_per_customer'] = data['revenue'] / (data['success_billing'] + 1)  # Avoid division by zero
data['average_order_value'] = data['revenue'] / (data['attempt_charging'] + 1)  # Avoid division by zero
data['inventory_turnover_ratio'] = data['revenue'] / (data['service_price'] + 1)  # Avoid division by zero


# Task 3: Comprehensive EDA

# Churn Risk Score Based on Engagement
data['churn_risk_score'] = (data['attempt_charging'] + data['success_billing']) / (data['attempt_charging'] + 1)  # Avoid division by zero
data['churn_risk_score'] = 1 - data['churn_risk_score']  # Higher engagement leads to lower risk score

# Compare churn risk across different countries
plt.figure(figsize=(12, 6))
sns.boxplot(x='country', y='churn_risk_score', data=data)
plt.title('Churn Risk Score Across Countries')
plt.xlabel('Country')
plt.ylabel('Churn Risk Score')
plt.xticks(rotation=45)
plt.show()

# Subscription Tenure
data['subscription_tenure_days'] = (data['renewal_date'] - data['subs_date']).dt.days

# Analyze how tenure correlates with revenue
plt.figure(figsize=(12, 6))
sns.scatterplot(x='subscription_tenure_days', y='revenue', data=data)
plt.title('Correlation Between Subscription Tenure and Revenue')
plt.xlabel('Subscription Tenure (Days)')
plt.ylabel('Revenue')
plt.show()

# Failed Charge Ratio
data['failed_charge_ratio'] = (data['attempt_charging'] - data['success_billing']) / (data['attempt_charging'] + 1)  # Avoid division by zero

# Analyze how this ratio impacts user retention
plt.figure(figsize=(12, 6))
sns.boxplot(x='failed_charge_ratio', y='status', data=data)  # Assuming 'status' indicates subscription status
plt.title('Impact of Failed Charge Ratio on Subscription Status')
plt.xlabel('Failed Charge Ratio')
plt.ylabel('Subscription Status')
plt.show()

# Visualize key insights
plt.figure(figsize=(12, 6))

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(0, inplace=True)

output_path = '/content/drive/MyDrive/MSIB Batch 7/Project Based Learning/cleaned_dataset_v1.csv'
data.to_csv(output_path, index=False)


print(f"Data cleaned, anomalies removed, features created, and scaled. Saved at {output_path}")
```
